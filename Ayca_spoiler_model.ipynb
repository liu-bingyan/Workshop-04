```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tokenizedMovieReviews = pd.read_json('tokenized.json')
```

```python
tokenizedMovieReviews
```

```python
vocab = {} 
count = 0


for lst in tokenizedMovieReviews['tokenized']: 
    for word in lst: 
        if word not in vocab: 
            vocab[word] = 1
        vocab[word] = vocab[word] + 1
print('vocabulary :', vocab) 
```

```python
count = 0
for word in vocab:
    if vocab[word] > 1000 : 
        count = count +1
        print(word)
        
```

```python
count
```

```python
actualVocab = {}
for word in vocab:
    if vocab[word] > 1000 : 
        actualVocab[word] = vocab[word]
```

```python
actualVocab
```

```python
tokenizedMovieReviews
```

```python
genres = {}
genreCount = 0
for lst in tokenizedMovieReviews['genre']: 
    for word in lst: 
        if word not in genres: 
            genres[word] = 1
            genreCount = genreCount + 1
        genres[word] = genres[word] + 1
print('genres :', genres) 
print(genreCount)
```

```python
for genre in genres:
    genreColumn = []
    for index, row in tokenizedMovieReviews.iterrows():
        genreColumn.append(row['genre'].count(genre))
    tokenizedMovieReviews[genre] = genreColumn
        
```

```python
tokenizedMovieReviews
```

```python
for word in actualVocab:
    wordColumn = []
    for index, row in tokenizedMovieReviews.iterrows():
        wordColumn.append(row['tokenized'].count(word))
    tokenizedMovieReviews[word] = wordColumn
        
```

```python
tokenizedMovieReviews = tokenizedMovieReviews.drop(columns=['tokenized', 'genre'])
```

```python
numericalSpoiler = []
for index, row in tokenizedMovieReviews.iterrows():
    if (row['is_spoiler'] == True):
        numericalSpoiler.append(1)
    else:
        numericalSpoiler.append(0)
tokenizedMovieReviews['is_spoiler'] = numericalSpoiler
```

```python
tokenizedMovieReviews = tokenizedMovieReviews.drop(columns=['user_id'])
```

```python
tokenizedMovieReviews
```

```python
y = tokenizedMovieReviews['is_spoiler']
X = tokenizedMovieReviews.drop(columns=['is_spoiler'])
```

```python
from sklearn.model_selection import train_test_split
X_train_test, X_validation, y_train_test, y_validation = train_test_split(X, y, test_size=0.2)
X_train, X_test, y_train, y_test = train_test_split(X_train_test, y_train_test, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import plot_confusion_matrix
confusion_matrix(y_test, y_pred)
plot_confusion_matrix(clf, X_test, y_test, cmap=plt.cm.Blues)
plt.show()
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
